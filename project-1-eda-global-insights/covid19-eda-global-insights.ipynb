```python
import pandas as pd          # for data handling
import matplotlib.pyplot as plt  # for plotting
import seaborn as sns        # for beautiful plots
```

```python
covid = r'covid.csv'
df=pd.read_csv(covid)

df.head() #show top 5 rows of the data set

df.info() #gives the information of datset

df.describe() #gives summary statistics
```

```python
df.isnull().sum() #checking for null values
```

```python
india_df = df[df['Country/Region'] == 'India'] # Filter for India

india_df = india_df.copy()
india_df['Date'] = pd.to_datetime(india_df['Date'])

# Plot new cases over time
plt.figure(figsize=(12, 6))
plt.plot(india_df['Date'], india_df['New cases'], label='New Cases', color='blue')
plt.title('Daily New COVID-19 Cases in India')
plt.xlabel('Date')
plt.ylabel('Number of New Cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and filter data
df = pd.read_csv('covid.csv')
india_df = df[df['Country/Region'] == 'India'].copy()
india_df['Date'] = pd.to_datetime(india_df['Date'])

# Plot daily deaths
plt.figure(figsize=(12, 6))
plt.plot(india_df['Date'], india_df['New deaths'], label='New Deaths', color='red')
plt.title('Daily New COVID-19 Deaths in India')
plt.xlabel('Date')
plt.ylabel('Number of New Deaths')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(india_df['Date'], india_df['New recovered'], label='New Recovered', color='green')
plt.title('Daily New COVID-19 Recoveries in India')
plt.xlabel('Date')
plt.ylabel('Number of New Recoveries')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
countries = ['India', 'France', 'Brazil']
multi_df = df[df['Country/Region'].isin(countries)].copy()
multi_df['Date'] = pd.to_datetime(multi_df['Date'])

plt.figure(figsize=(12, 6))
for country in countries:
    country_df = multi_df[multi_df['Country/Region'] == country]
    plt.plot(country_df['Date'], country_df['New cases'], label=country)

plt.title('Daily New COVID-19 Cases Comparison')
plt.xlabel('Date')
plt.ylabel('Number of New Cases')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(india_df['Date'], india_df['Confirmed'], label='Total Confirmed Cases', color='purple')
plt.plot(india_df['Date'], india_df['Deaths'], label='Total Deaths', color='red')
plt.plot(india_df['Date'], india_df['Recovered'], label='Total Recovered', color='green')

plt.title('Cumulative COVID-19 Cases, Deaths, and Recoveries in India')
plt.xlabel('Date')
plt.ylabel('Count')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
